## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

```python
sample_json_str = json.load((open('data/world_bank_projects.json')))
json_normalize(sample_json_str,'mjsector_namecode').name.value_counts()
```

```python
pd.set_option('display.max_rows', 1000)
sample_json_str = json.load((open('data/world_bank_projects.json')))
df = json_normalize(sample_json_str,'mjsector_namecode')
print df.code.count()
print df.name.count()
# It seems that there are no missing values in the name columns?
# for the sake of practice, we would generate some random NA and fill it in
```

```python
import numpy as np
rand = np.random.randint(0,1405,20)
df.name.iloc[rand] = np.nan
df.name.count()
```

```python
#construct a lookup table for code&name
code_lookup = json_normalize(sample_json_str,'mjsector_namecode').groupby(['code','name']).size()
code_lookup = code_lookup.reset_index().rename(columns={0:'count'})
code_lookup2 = code_lookup[['code','name']]#.set_index('code')
code_lookup2 = pd.Series(list(code_lookup['name']), code_lookup.code).rename('name')
code_lookup2
df.loc[df['name'].isnull(),'name']= df['code'].map(code_lookup2)
df.iloc[rand]
```
