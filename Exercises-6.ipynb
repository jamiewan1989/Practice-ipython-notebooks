### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
df1 = sales1.rename(columns={
    'Book title': 'title',
    'Number sold': 'number',
    'Sales price': 'their_price',
    'Royalty paid': 'we_got',
    })
df1['currency'] = 'USD'
df1
sales2.head()
```

```python
df2 = sales2.rename(columns = {
        'Title': 'title',
        'Units sold': 'number',
        'List price': 'their_price',
        'Royalty': 'we_got',
    })
df2 = df2.fillna(" ")
df2['currency'] = df2.title.str.extract(r'\((.*?)\)')
df2 = df2.fillna(method = 'bfill')
df2 = df2[df2.number != " "]
df2
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
df['total_paid'] = df.their_price * df.number
df.groupby(['title', 'currency'])[['total_paid']].sum()
```
